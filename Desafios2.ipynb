```python
import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
cod_compostos
```

```python
ordem_composto = dados.query('composto in @cod_compostos')['composto'].value_counts().index
```

##DESAFIO 01 e 02

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(12, 8))
sns.set(font_scale=1.5)
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'),order=ordem_composto)
ax.set_title('Top 5 compostos')
plt.show()
```

```python
plt.figure(figsize=(10,7))
ax = sns.histplot(data=dados,x='g-25', kde=True)
plt.show()
```

##DESAFIO 04

O .describe(), e o .info(), são métodos utilizados para obter informações do dataframe.

count = Conta a quantidade de linha que se encontra na coluna 

mean = Mostra a media da coluna

std =  Desvio padrão dos dados apresnetados na coluna.

min = mostra o valor mínimo da coluna.

25% = Mostra o primeiro quarto de dados.

50% =  Mostra onde se encontra a metade dos dados na distribuição

75% = Mostra onde está os 75% do dados na distribuição.

max = Mostra o valor máximo do contjunto de dados na coluna

##DESAFIO 05

É importante deixarmos os graficos "bonitão" pois facilita que pessoas que não entendem de dados ou oque exatamente esta sendo tratado, com os graficos fica mais facil para uma pessoa se localizar e entender oque esta sendo tratado.

##DESAFIO 06

```python
fig = plt.figure(figsize=(15,15))
sns.boxplot(x="tempo", y='g-0', hue='tratamento', data = dados, ax=fig.add_subplot(211))
plt.xlabel('Tempo', weight = 'bold')
sns.boxplot(x="tempo", y='g-1', hue='tratamento', data = dados, ax=fig.add_subplot(212))
plt.xlabel('Tempo', weight = 'bold')
plt.show()
```
